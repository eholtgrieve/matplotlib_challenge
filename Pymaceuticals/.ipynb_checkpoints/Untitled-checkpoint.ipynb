```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"
```

```python
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
mouse_study = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")
mouse_study.head()
```

```python
mouse_metadata['Mouse ID'].count()
```

```python
duplicate_id = mouse_study.loc[mouse_study.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_id
```

```python
duplicate_df = mouse_study.loc[mouse_study["Mouse ID"] == "g989", :]
duplicate_df
```

```python
mouse_study = mouse_study[mouse_study['Mouse ID'].isin(duplicate_id)==False]
mouse_study.head()
```

```python
len(mouse_study['Mouse ID'].unique())
```

```python
drug_regimen = mouse_study.groupby("Drug Regimen")

mean = drug_regimen["Tumor Volume (mm3)"].mean()
median = drug_regimen["Tumor Volume (mm3)"].median()
variance = drug_regimen["Tumor Volume (mm3)"].var()
std = drug_regimen["Tumor Volume (mm3)"].std()
sem = drug_regimen["Tumor Volume (mm3)"].sem()

drug_summary = pd.DataFrame([mean, median, variance, std, sem]).T
drug_summary.columns = ["Mean Tumor Volume","Median Tumor Volume","Tumor Volume Variance", 
                        "Tumor Volume Std. Dev.", "Tumor Volume Std. Err."]
    

drug_summary
```

```python
drug_summary2 = {"Tumor Volume (mm3)": ["mean", "median", "variance", "std", "sem"]}
mouse_study.groupby('Drug Regimen').agg(drug_summary2)
```
